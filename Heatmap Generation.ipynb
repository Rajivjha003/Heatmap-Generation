### Import Libraries

```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Dataset Loading

```python
# Loading Dataset

# Defining the size of grid(100 randomly)
grid_size = 100, 100

# Creating a grid b/w 0-100
thermal_data = np.random.uniform(low=0, high=100, size=grid_size)
```

Im generating the random thermal data. This could be done using a random number generator to simulate temperature readings and data is structured in 2d array to represent a grid of temperature reading.

###  Data Transformation

### Data Scaling

```python
# Normalize the data
normalized_data = (thermal_data - np.min(thermal_data)) / (np.max(thermal_data) - np.min(thermal_data))
```

Here we normalize the thermal data in range of (0-1) using min-max scaler. this also help in ensuring consistent color mapping in the heatmap.

```python
# Now defining a threshold for 'hot' areas
threshold = 0.8

# Identify hot areas
hot_areas = normalized_data > threshold
```

The threshold is set to 0.8, above threshold limit considering "hot. We will use the threshold value later for showing contour lines.

## ***Data Vizualization***

#### Chart 

```python
# Defining  a function to generate heatmap from 2D to thermal data
def generate_heatmap(data, cmap='viridis', alpha=1.0, show_contour=False):
    plt.imshow(data, cmap=cmap, interpolation='nearest', alpha=alpha)
    plt.colorbar()
    if show_contour:
        plt.contour(data > threshold, colors='red', linestyles='solid', linewidths=2)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Thermal Data Heatmap')
    plt.show()
```

This function utilizes Matplotlib to create a heatmap, offering flexibility in color mapping through the cmap parameter.Optionally, contour lines can be shown for areas with temperatures exceeding a predefined threshold. Axes are labeled, and a title is added for better interpretation.
The function does not return anything, as its primary purpose is to visualize the thermal data.


```python
# Generate and display the heatmap
generate_heatmap(normalized_data, cmap='viridis', alpha=0.8, show_contour=True)
```
